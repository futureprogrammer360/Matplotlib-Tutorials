# Histograms

Histograms are used to show the frequency (number of observations) of data in successive numerical intervals of equal size.

Similar to bar graphs, histograms use bars to show frequency. The difference between the two types of graphs is that while bar graphs visualize **categorical data**, histograms are used for **numerical data**.

For instance, bar graphs can show the number of programmers in different companies (company A, company B, company C, etc), histograms can show the number of programmers in different ages using intervals (11-20, 21-30, 31-40, etc).

```python
import matplotlib.pyplot as plt
```

`matplotlib.pyplot.hist()` is Matplotlib's histogram-making function. We can just pass a list of numbers into the function call, and Matplotlib will plot a histogram for us.

```python
import random

x = [random.randint(0, 50) for i in range(500)]
plt.hist(x)
plt.show()
```

`plt.hist()` takes another optional parameter, `bins`, which is the number of bins in the histogram.

```python
x = [random.randint(0, 50) for i in range(500)]
plt.hist(x, bins=25)
plt.show()
```

We can create a normal distribution using `numpy.random.normal()`.

`numpy.random.normal(loc, scale, size)` takes
* `loc`: the center of distribution
* `scale`: spread of distribution in standard deviation
* `size`: number of values in distribution

Since `numpy.random.normal()` produces a normal distribution, the histogram plotted with that data should be approximately bell-shaped.

```python
import numpy as np

x = np.random.normal(0, 1, 100000)
plt.hist(x, bins=100)
plt.show()
```

We can also plot multiple histograms on one graph to compare two groups.

```python
junior_programmers = np.random.normal(3, 2, 200)
senior_programmers = np.random.normal(6, 3, 120)

plt.hist(junior_programmers, label="Junior Programmers")
plt.hist(senior_programmers, label="Senior Programmers")
plt.title("Number of Programming Languages Learned")
plt.legend()
plt.show()
```
