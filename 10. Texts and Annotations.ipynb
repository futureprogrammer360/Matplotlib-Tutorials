# Texts and Annotations

Sometimes, it can be helpful to include some texts and annotations in our data visualizations to better tell the story behind our graphs and plots.

````python
import numpy as np
import matplotlib.pyplot as plt
````

## Text

To add text somewhere on our plot, we can use `matplotlib.pyplot.text()`, which takes the following parameters:

* `x`: x-coordinate of text position
* `y`: y-coordinate of text position
* `s`: text

````python
plt.plot(np.arange(-5, 5, 0.1), -np.arange(-5, 5, 0.1) ** 2)
plt.text(0, -10, "A parabola!")
plt.show()
````

We can change alignment of the text using the `horizontalalignment` and `verticalalignment` parameters.

````python
plt.subplot(1, 2, 1)
plt.plot(range(3), range(3))
plt.text(1, 1, "Positive slope")

plt.subplot(1, 2, 2)
plt.plot(range(3), list(range(3))[::-1])
plt.text(1, 1, "Negative slope",
         horizontalalignment="center", verticalalignment="center")

plt.show()
````

We can also change other text properties like size, color, rotation, font family, etc.

````python
plt.figure(figsize=(6, 2))
plt.plot(range(10), [1 for _ in range(10)], ls="none")

plt.text(0, 1, "ABC", fontsize="xx-large")
plt.text(2, 1, "BCD", color="red")
plt.text(4, 1, "CDE", rotation=60)
plt.text(6, 1, "DEF", fontfamily="georgia")
plt.text(8, 1, "EFG", alpha=0.2)

plt.show()
````

## Annotations

Annotations are similar to texts, but gives us the option of having an arrow that points to another point.

```python
plt.annotate(text, xy, xytext, arrowprops)
```

* `text`: text displayed
* `xy`: coordinates of annotated point
* `xytext`: coordinates of text
* `arrowprops`: properties of arrow (optional)

````python
plt.plot(np.arange(-5, 5, 0.1), -np.arange(-5, 5, 0.1) ** 2)
plt.annotate("Maximum of $f(x)=x^2$", xy=(0, 0), xytext=(-1, -15),
            arrowprops={"arrowstyle": "->"}, horizontalalignment="center")
plt.show()
````

Other annotation customizations can be found here: https://matplotlib.org/stable/tutorials/text/annotations.html
