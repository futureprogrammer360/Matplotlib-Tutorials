# Titles, Labels, Legends

So far, our plots only consists of visualized data: points and lines. However, plots such as the one below are not very useful.

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(11))
y = [2 ** n for n in x]
plt.plot(x, y)
plt.show()
```

We don't know what the data is about: the plot above lacks context. It's always a good idea to include labels of our data so viewers know what we are graphing.

## Titles

Titles are used to indicate what the plot is all about. The `plt.title()` function sets the title for us.

```python
x = list(range(11))
y = [2 ** n for n in x]
plt.plot(x, y)

plt.title("An exponential function")

plt.show()
```

## Labels

We can also use the `plt.xlabel()` and `plt.ylabel()` functions to set axis labels.

```python
x = list(range(11))
y = [2 ** n for n in x]
plt.plot(x, y)
plt.title("Human Population Simulation")

plt.xlabel("Time")
plt.ylabel("Population (thousands)")

plt.show()
```

## Legends

Sometimes, we have multiple lines on our plot. It's very difficult to know which line is which data, so it's helpful to label the data using legends.

The `plt.plot()` method takes an argument called `label`, which is used to label that particular line. Using the `plt.legend()` method, we can then display the labels.

```python
x = list(range(6))
plt.plot(x, [2 ** n for n in x], "-b", label="$2^x$")
plt.plot(x, [3 ** n for n in x], "--r", label="$3^x$")

plt.legend()

plt.show()
```

Alternatively, we can pass the labels directly into `plt.legend()` as a list. The order must match the `plt.plot()` calls.

```python
x = list(range(6))
plt.plot(x, [2 ** n for n in x], "-b")
plt.plot(x, [3 ** n for n in x], "--r")

plt.legend(["$2^x$", "$3^x$"])

plt.show()
```
