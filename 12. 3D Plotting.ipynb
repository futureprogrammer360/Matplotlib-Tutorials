# 3D Plotting

In addition to the wide range of data visualizations it supports in 2 dimensions, Matplotlib also allows us to create 3D plots, which have 3 axes - x, y, and z.

```python
import numpy as np
import matplotlib.pyplot as plt
```

When creating 3D plots, we need to create a `pyplot.Figure` object with an `Axes` whose `projection` is set to `"3d"`. This can be done as shown below.

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
```

#### 3D plots shown in Jupyter Notebook are not interactive. To allow operations like zooming and panning, you can write and run the code in a `.py` file. Matplotlib 3D plots opened by a script adds user interactivity.

## 3D Scatterplots

A common plot in 3D is scatterplots. The only differences 3D scatterplots have from 2D scatterplots are `projection="3d"` and the addition of a z-axis.

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

ax.scatter(range(10), range(10), range(10), marker="x", c="g")
ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_zlabel("z values")
ax.set_title("A 3D Scatterplot")

plt.show()
```

## 3D Line Plots

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

z = np.linspace(0, 75, 1000)
x = np.cos(z)
y = np.sin(z)
ax.plot(x, y, z)

plt.show()
```

## Multiple Plots

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

z = np.linspace(0, 20, 1000)
x = np.cos(z)
y = np.sin(z)
ax.plot(x, y, z, "red")

x += 0.1 * np.random.randn(1000)
y += 0.1 * np.random.randn(1000)
ax.scatter(x, y, z, c=z)

plt.show()
```
