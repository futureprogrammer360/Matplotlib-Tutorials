# Stack Plots and Fill Between

Stack plots (also known as area plots) are a special type of line graphs where the area under each line is filled. Although a seemingly minor change, stack plots are much better in some scenarios.

In certain cases, we not only want to show changes of some data over time in categories, but we also want to show the change in the difference between the categories. For instance, a line graph can show the increase in ownership of both gasoline-powered and electric vehicles. But a stack plot is better suited at showing how although gas-powered cars are more popular, the gap is closing.

In the second half of the lesson, we will talk about how we can fill in the area between two lines in matplotlib.

````python
import matplotlib.pyplot as plt
````

We use `plt.stackplot()` to make stackplots. There are two ways to pass in the data.

```python
plt.stackplot(x, y)              # y is 2d array of the y-values
plt.stackplot(x, y1, y2, y3...)  # y1, y2, y3, etc are 1d arrays
```

````python
x = list(range(10))
y1 = [1, 2, 4, 3, 5, 6, 7, 9, 8, 11]
y2 = [2, 3, 2, 6, 3, 5, 9, 6, 7, 9]
y3 = [5, 4, 6, 4, 6, 7, 8, 16, 19, 18]
y4 = [0, 0, 0, 2, 5, 7, 12, 17, 27, 34]

plt.stackplot(x, y1, y2, y3, y4)
plt.show()
````

````python
x = list(range(10))
y1 = [1, 2, 4, 3, 5, 6, 7, 9, 8, 11]
y2 = [2, 3, 2, 6, 3, 5, 9, 6, 7, 9]
y3 = [5, 4, 6, 4, 6, 7, 8, 16, 19, 18]
y4 = [0, 0, 0, 2, 5, 7, 12, 17, 27, 34]

y = [y1, y2, y3, y4]

plt.stackplot(x, y)
plt.show()
````

## Labels and Legends

`plt.stackplot()`'s `labels` parameter specifies the labels for each "stack".

````python
x = list(range(10))
y1 = [1, 2, 4, 3, 5, 6, 7, 9, 8, 11]
y2 = [2, 3, 2, 6, 3, 5, 9, 6, 7, 9]
y3 = [5, 4, 6, 4, 6, 7, 8, 16, 19, 18]
y4 = [0, 0, 0, 2, 5, 7, 12, 17, 27, 34]

plt.stackplot(x, y1, y2, y3, y4, labels=["A", "B", "C", "D"])
plt.title("Stackplot")
plt.xlabel("Time")
plt.ylabel("Number of things")
plt.legend(loc="upper left")
plt.show()
````

# Fill Between

Another way of creating area plots is using `matplotlib.pyplot.fill_between()`, which fills the area between 2 lines.

````python
import random

x = range(150)
y1 = [a + random.randint(10, 25) for a in x]
y2 = [a + random.randint(10, 25) for a in y1]

plt.plot(x, y1)
plt.plot(x, y2)

plt.fill_between(x, y1, y2, alpha=0.5, facecolor="green")

plt.show()
````
