# Scatter Plots

Scatter plots are visualizations with points that illustrate the relationship between variables. Each dot represents a value in the dataset.

```python
import matplotlib.pyplot as plt
```

We use `plt.scatter()` to draw scatter plots. Similar to `plt.plot()`, we pass in lists for `x` and `y`.

```python
import numpy as np

x = np.random.normal(0, 1, 750)
y = np.random.normal(0, 1, 750)

plt.scatter(x, y)
plt.show()
```

```python
import random

# Pick 100 numbers from 0 to 99
x = random.choices(range(100), k=100)
# Create a mostly linear relationship
y = [a + random.randint(-15, 15) for a in x]

plt.scatter(x, y)

x = random.choices(range(100), k=100)
y = [a + random.randint(-50, 50) for a in x]

plt.scatter(x, y)

plt.show()
```

## Customizations

The advantage that `plt.scatter()` has over `plt.plot()` is that `plt.scatter()` allows us to customize attributes for each individual dot.

### Colors

We can change the color for all dots. We can also change it for each dot using the `c` parameter. The argument we pass is a list of values the same length as `x` and `y`. Each color element corresponds to a dot.

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

colors = ["red", "green", "blue"]
plt.scatter(x, y, c=random.choices(colors, k=len(x)))
plt.show()
```

## Size

Changing size is similar to color. We pass in a list of sizes for parameter `s`.

```python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
sizes = [a ** 4 + b ** 4 for a, b in zip(x, y)]

plt.scatter(x, y, color="orange", s=sizes)
plt.show()
```

## Transparency

The `alpha` parameter modifies the transparency of each dot, taking a value between 0 and 1 (0 is transparent, 1 is opaque).

```python
x = np.random.normal(0, 1, 250)
y = np.random.normal(0, 1, 250)

plt.scatter(x, y, alpha=[random.random() for _ in x])
plt.show()
```

## Other customizations

As you might expect, we can also change things like marker style.

```python
x = np.random.normal(0, 1, 250)
y = np.random.normal(0, 1, 250)

plt.scatter(x, y, marker="X", edgecolors="red", linewidths=0.5)
plt.show()
```

## A final scatter plot

Let's combine all the modifications into one scatter plot.

```python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

sizes = [0.05 + a ** 4 + b ** 4 for a, b in zip(x, y)]
colors = [(random.random(), random.random(), random.random()) for _ in x]
alphas = [max(random.random(), 0.25) for _ in x]

plt.scatter(x, y, c=colors, s=sizes, alpha=alphas, edgecolors=colors[::-1])
plt.show()
```
