# Object-Oriented Interface

So far, we have been using `matplotlib`'s **procedural/functional interface** (`plt` function calls) to create visualizations. This style of `matplotlib` was made to imitate the programming language MATLAB.

`matplotlib` also has an **object-oriented interface** using method calls on `Axes` and `Figure`s which can sometimes make our code more readable and intuitive.

Before we see how `matplotlib`'s object-oriented code look like, let's first understand what the objects in a `matplotlib`'s plot are.

![Anatomy of a figure](https://matplotlib.org/stable/_images/anatomy.png)

Whichever approach we choose to use - functional or object-oriented - our graphs are generated on `Figure`s. `Figure`s contain everything our graphs have. Each `Figure` can contain one or more `Axes`, which is essentially a region for plotting. An `Axes` then contains the x-axis, y-axis, markers, titles, plots, texts, etc.

In the diagram above, everything that is labeled `ax.something` is part of the `Axes`, which is part of the `Figure`.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.1)
```

```python
# Functional style
plt.plot(x, 1 * x ** 2, label="$y=1x^2$")
plt.plot(x, 2 * x ** 2, label="$y=2x^2$")
plt.plot(x, 3 * x ** 2, label="$y=3x^2$")
plt.title("Some quadratic functions")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
plt.show()
```

```python
# Object-oriented style
fig = plt.figure()
ax = fig.subplots()
ax.plot(x, 1 * x ** 2, label="$y=1x^2$")
ax.plot(x, 2 * x ** 2, label="$y=2x^2$")
ax.plot(x, 3 * x ** 2, label="$y=3x^2$")
ax.set_title("Some quadratic functions")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend()
plt.show()
```

### Advantages of object-oriented interface

While the functional `pyplot` interface is sufficient for most cases, the object-oriented approach offers some advantages in some situations.

The functional interface is also known as the *implicit* interface, because `matplotlib` keeps track of the current figure and axes, and basically "guesses" which ones the programmer wants a plot to go on.

This works most of the time, but if in programs with many many figures and axes, the implicit interface can create a mess. The OO approach is also called the *explicit* interface, because we specify exactly which figure and axes we are plotting on.

For more information, see the official documentation: https://matplotlib.org/stable/users/explain/api_interfaces.html
