# Bar Graphs

Bar graphs are used to compare data in distinct categories using bars.

````python
import matplotlib.pyplot as plt
````

Matplotlib's `plt.bar()` function is used to create bar graphs. Here's the most basic syntax:

```python
plt.bar(x, height)
```

* `x`: "x coordinates of the bars", which is basically the x labels (usually the category labels)
* `height`: heights of the bars

````python
plt.bar([1, 2, 3, 4, 5, 6], [1, 4, 9, 16, 25, 36])
plt.show()
````

````python
plt.bar(["A", "B", "C", "D", "E", "F"], [1, 4, 9, 16, 25, 36])
plt.show()
````

Here's a few other customizations you can add to your bar graphs.

````python
plt.bar(
    ["Java", "C++", "JavaScript", "Python"],
    [1, 4, 9, 16],
    width=1,  # default is 0.8
    color="orange",
    edgecolor="red"
)
plt.title("Favorite subjects")
plt.show()
````

`plt.bar()` returns a `BarContainer` object. We can pass this object into `plt.bar_label()` to label the value of a bar.

````python
container = plt.bar(["A", "B", "C", "D", "E", "F"], [1, 4, 9, 16, 25, 36])
plt.bar_label(container)
plt.show()
````

## Stacked bar graphs

The `bottom` argument allows us to create stacked bar graphs. We just need to call `plt.bar()` multiple times, and use the `bottom` argument to stack the top bars on top of the bottom ones.

````python
labels = ["Q1", "Q2", "Q3", "Q4"]
food = [50, 60, 56, 49]
transportation = [24, 32, 29, 31]
electricity = [10, 8, 11, 16]

c1 = plt.bar(labels, food, label="Food")
c2 = plt.bar(labels, transportation, label="Transportation", bottom=food)
c3 = plt.bar(labels, electricity, label="Electricity",
        bottom=[f + t for f, t in zip(food, transportation)])

plt.bar_label(c1)
plt.bar_label(c2)
plt.bar_label(c3)

plt.title("Expenses")
plt.xlabel("Quarters")
plt.ylabel("Spendings ($)")
plt.legend()
plt.show()
````

## Double bar charts

Double bar graphs are easy to make using `plt.bar()`.

````python
labels = ["Spring", "Summer", "Fall", "Winter"]
x = range(len(labels))
width = 0.35

low_temps = [8, 19, 9, 1]
high_temps = [19, 38, 16, 11]

c1 = plt.bar([s - width / 2 for s in x],
             low_temps, width, label="Low Temperatures")
c2 = plt.bar([s + width / 2 for s in x],
             high_temps, width, label="High Temperatures")

plt.bar_label(c1)
plt.bar_label(c2)
plt.title("Seasonal Temperatures")
plt.xlabel("Season")
plt.ylabel("Temperature")
plt.xticks(x, labels)
plt.legend()

plt.show()
````

## Horizontal bar graphs

`plt.barh()` is used to make horizontal bar graphs in matplotlib. Here's the basic syntax:

```python
plt.barh(y, width)
```

* `y` is the y-coordinates of the bars
* `width` is the widths of the bars

These two parameters effectively corresponds to the `x` and `height` arguments of `plt.bar()`.

There are also optional parameters such as `height` and `color`.

````python
plt.barh(["Google", "Instagram", "Facebook", "YouTube"],
         [1, 1.44, 2.92, 2.5], height=0.7,
         color=["green", "purple", "blue", "red"])
plt.show()
````
