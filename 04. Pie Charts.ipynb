# Pie Charts

A very useful type of charts in data science is the pie chart, which is used to display division of a total into categories, showing part-to-whole relationships.

```python
import matplotlib.pyplot as plt
```

Matplotlib allows us to easily create and customize pie charts using the `plt.pie()` function, which takes a parameter `x`: the sizes of the pie pieces as a list. And just like `plt.plot()`, we need to call `plt.show()` to display the plot.

```python
plt.pie([1, 2, 3, 4, 5])
plt.show()
```

Right now, we have no clue what the sectors are representing. The `labels` parameter lets us display labels on the pie chart.

```python
plt.pie(
    [1, 2, 3, 4, 5],
    labels=["History", "English", "Science", "Math", "Programming"]
)
plt.show()
```

There're also a lot of other customizations you can add to your pie chart. Below is a plot with some of these modifications.

```python
plt.pie(
    [1, 2, 3, 4, 5],
    labels=["History", "English", "Science", "Math", "Programming"],
    autopct="%.1f%%",
    explode=[0, 0, 0, 0, 0.1],
    colors=["#4B89B8", "#3D9BC7", "#2FACD6", "#20BEE4", "#12CFF3"],
    shadow=True
)
plt.title("Favorite subjects")
plt.show()
```
