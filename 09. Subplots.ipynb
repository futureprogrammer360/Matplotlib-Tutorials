# Subplots

Matplotlib allows us to graph multiple plots in one figure. The `pyplot` module supplies the `subplot()` function for this purpose.

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 300  # Increase plot resolution
```

`plt.subplot()` requires a few numbers:

* Number of rows in figure
* Number of columns in figure
* Position of the current plot

There are 2 ways to specify these arguments.
* `plt.subplot(nrows, ncols, index)` (example: `plt.subplot(1, 2, 1)`)
* `plt.subplot(nrowsncolsindex)` (example: `plt.subplot(121)`)

For instance, the code below creates a figure with 1 row and 2 columns (side-by-side subplots).

```python
x = list(range(10))

plt.subplot(1, 2, 1)
plt.plot(x, x)

plt.subplot(1, 2, 2)
plt.plot(x, list(reversed(x)))

plt.show()
```

To create top-down subplots we only need to change the input to the `subplot()` function call.

```python
x = list(range(5))

plt.subplot(211)
plt.plot(x, x)

plt.subplot(212)
plt.plot(x, list(reversed(x)))

plt.show()
```

And of course, we can have more than just two plots in one figure. If we have more than 9 though, we will need to use the `plt.subplot(nrows, ncols, index)` syntax.

```python
for i in range(1, 10):
    plt.subplot(3, 3, i)
    plt.plot(range(10), [i * x for x in range(10)])
    plt.xlim([0, 9])
    plt.ylim([0, 81])

plt.tight_layout()
plt.show()
```

In the code above, we call `plt.tight_layout()` to adjust subplot params so that our subplots fit in the figure more cleanly.

### Setting subplot properties

We can set subplot properties like titles, legends, etc, just as we normally would.

```python
import math
import string
import numpy as np

plt.subplot(2, 1, 1)
x = range(-50, 50)
plt.plot(x, [math.sin(0.1 * n) for n in x], label="$sin(x)$")
plt.plot(x, [math.cos(0.1 * n) for n in x], label="$cos(x)$")
plt.legend(loc="upper right")
plt.title("Trigonometric functions")

plt.subplot(2, 2, 3)
plt.hist(np.random.normal(0, 1, 100000), 100)
plt.title("A normal distribution")

plt.subplot(2, 2, 4)
plt.pie(range(1, 6), labels=string.ascii_uppercase[:5], autopct="%d%%")
plt.title("A pie chart")

plt.tight_layout()
plt.show()
```

#### `suptitle()`

We can set a "suptitle" to a figure, which is a centered title of all the plots in a figure, using `plt.suptitle()`.

```python
x = range(-50, 50)

plt.subplot(2, 1, 1)
plt.plot(x, [math.sin(0.1 * n) for n in x])
plt.title("$sin(x)$")

plt.subplot(2, 1, 2)
plt.plot(x, [math.cos(0.1 * n) for n in x])
plt.title("$cos(x)$")

plt.suptitle("Sine and Cosine")

plt.tight_layout()
plt.show()
```
