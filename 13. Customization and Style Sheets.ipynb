# Customization and Style Sheets

Matplotlib allows us to change its configuration and plotting styles to customize our data visualizations in 3 ways:

1. `rcParams`
2. Style sheets
3. Modifying the `matplotlibrc` file

We will be discussing the first 2 methods, since it is rarely necessary to modify the `matplotlibrc` file.

```python
import numpy as np
import matplotlib as mpl         # Notice matplotlib package imported
import matplotlib.pyplot as plt  # In addition to its pyplot module
```

```python
def plotting_function():
    for i in range(1, 10):
        x = np.arange(-10, 10, 0.01)
        y = np.cos(x) * i
        plt.plot(x, y)
    plt.show()

plotting_function()
```

## `rcParams`

`rc` stands for runtime configuration. We can change these configurations using the `matplotlib.rcParams` variable.

```python
print(len(mpl.rcParams))
print()
print(mpl.rcParams)
```

The `matplotlib.rcParams` variable is like a dictionary, so we can access and modify the params directly using dictionary syntax.

```python
print(mpl.rcParams["lines.linewidth"])
plotting_function()
```

```python
mpl.rcParams["lines.linewidth"] = 5
print(mpl.rcParams["lines.linewidth"])
plotting_function()
```

`plt.rcdefaults()` can be used to reset `rcParams` to defaults.

```python
plt.rcdefaults()

plotting_function()
```

## Style sheets

Style sheets are pre-defined `rcParams` that can be imported in our own programs. Matplotlib provides many styles that can be loaded with `plt.style.use()`.

```python
print(plt.style.available)
```

Style sheets reference: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

```python
plt.style.use("ggplot")

plotting_function()
```
