## Basic plot

Now let's learn how to create a static plot using the `matplotlib` package.

### Import

The `pyplot` module of `matplotlib` provides the basic graphing utilities we'll need. This module is typically imported under the alias `plt`.

````python
import matplotlib.pyplot as plt
````

### Simple plot

The `plt.plot()` function is used to create simple plots. Here is the basic function call signature.

```python
plt.plot(x, y)
```

After we create the plot, we can display it using `plt.show()`.

````python
x = [1, 2, 3, 4, 5]
y = [4, 3, 1, 2, 6]
plt.plot(x, y)

plt.show()
````

````python
x = range(-100, 101)
y = [n ** 2 for n in x]
plt.plot(x, y)

plt.show()
````

You can also graph multiple lines by calling `plt.plot()` multiple times.

````python
import math

x = range(-100, 101)
y = [math.sin(0.1 * n) for n in x]
plt.plot(x, y)

x = range(-100, 101)
y = [math.cos(0.1 * n) for n in x]
plt.plot(x, y)

plt.show()
````
