# Plot Styles

In the last lesson, you created a few simple line plots using `matplotlib.pyplot.plot()`. Today, you will learn how to modify plot styles.

````python
import matplotlib.pyplot as plt
````

By default, a matplotlib plot looks like this:

````python
plt.plot([1, 2], [1, 2])
plt.show()
````

## Customizing color

The `color` and `c` parameters do the same thing: modify the color of the plot.

````python
plt.plot([1, 2], [1, 2], c="red")
plt.plot([1, 2], [2, 1], color="#593F62")
plt.show()
````

## Customizing linestyle

The line style of plots can also be changed depending on our needs.

The `linestyle` or `ls` parameters accept the following values according to matplotlib's documentation:

| linestyle                          | description      |
|------------------------------------|------------------|
| `'-'` or `'solid'`                 | solid line       |
| `'--'` or `'dashed'`               | dashed line      |
| `'-.'` or `'dashdot'`              | dash-dotted line |
| `':'` or `'dotted'`                | dotted line      |
| `'none'`, `'None'`, `' '`, or `''` | draw nothing     |

````python
plt.plot([1, 2], [5, 5], linestyle="-")
plt.plot([1, 2], [4, 4], linestyle="--")
plt.plot([1, 2], [3, 3], linestyle="-.")
plt.plot([1, 2], [2, 2], linestyle=":")
plt.plot([1, 2], [1, 1], linestyle="none")

plt.show()
````

## Customizing markers

Matplotlib also allows us to customize the markers in a plot.

Here's the link to the full list of options: https://matplotlib.org/stable/api/markers_api.html

Below are some of the valid values.

````python
plt.plot([1, 2, 3], [4, 4, 4], marker=".")
plt.plot([1, 2, 3], [3, 3, 3], marker="D")
plt.plot([1, 2, 3], [2, 2, 2], marker="*")
plt.plot([1, 2, 3], [1, 1, 1], marker="s")

plt.show()
````

You can also change other attributes of the markers, such as their size, color, etc.

## `fmt`

We can also pass in a single format string into the `plot()` method call to customize the marker, line, and color of the plot in one go.

Below is the order of the values in the string.

```python
fmt = '[marker][linestyle][color]'
```

For example, if you want to make a plot with
* marker: square
* linestyle: dashed
* color: red

You can use the `fmt` string of `"s--r"` (square markers, dashed line, red color)

````python
import random

x = range(50)
y = random.sample(list(range(50)), k=50)
plt.plot(x, y, "s--r")
plt.show()
````
